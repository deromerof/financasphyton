```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt 
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI', '^BVSP']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='1997-1-1') ['Adj Close']
```

```python
ind_data.info()
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

^é utlizado para indicar que trata-se de índice de ações
GSPC = SP500, IXIC = NADASQ, GDAXI = DAX, FTSE = NASDAQ

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(15, 6));
plt.show()
```

```python
ind_returns = (ind_data/ind_data.shift(1))-1
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean()*250
annual_ind_returns
```
